```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import sklearn
```

# Importar dados

```python
df_ddos = pd.read_csv('data/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
df_ps = pd.read_csv('data/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv')
```

```python
df_ddos.head(10)
```

```python
df_ps.head(10)
```

```python
## concateraremos os datasets, para criarmos um dataset √∫nico com tipos variados de ataques
```

```python
df_mix = pd.concat([df_ddos,df_ps])
```

```python
df_mix.info()
```

```python
df_mix = df_mix.sample(512212)
```

```python
df_mix.head(10)
```

```python
df_agent1 = df_mix.iloc[0:50000].copy()
df_agent2 = df_mix.iloc[50000:100000].copy()
df_validation = df_mix.iloc[100000:150000].copy()
```

```python
df_agent1 = df_agent1.reset_index(drop=True)
df_agent2 = df_agent2.reset_index(drop=True)
df_validation = df_validation.reset_index(drop=True)
```

```python
df_agent2.head()
```

```python
# Remover espa√ßos extras
df_agent1.columns = [col.strip() for col in df_agent1.columns]  
df_agent2.columns = [col.strip() for col in df_agent2.columns]
df_validation.columns = [col.strip() for col in df_validation.columns]
```

```python
benign = (df_agent1['Label'] == "BENIGN").sum()
ddos = (df_agent1['Label'] == "DDoS").sum()
portscan = (df_agent1['Label'] == "PortScan").sum()
print(f"benign: {benign}, ddos: {ddos}, portscan: {portscan}")
```

```python
benign = (df_agent2['Label'] == "BENIGN").sum()
ddos = (df_agent2['Label'] == "DDoS").sum()
portscan = (df_agent2['Label'] == "PortScan").sum()
print(f"benign: {benign}, ddos: {ddos}, portscan: {portscan}")
```

```python
benign = (df_validation['Label'] == "BENIGN").sum()
ddos = (df_validation['Label'] == "DDoS").sum()
portscan = (df_validation['Label'] == "PortScan").sum()
print(f"benign: {benign}, ddos: {ddos}, portscan: {portscan}")
```

# Tratamento de dados

```python
df_agent1.isnull().sum()
```

```python
df_agent2.isnull().sum()
```

```python
df_validation.isnull().sum()
```

```python
# Replace inf and -inf with NaN
df_agent1.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# Replace inf and -inf with NaN
df_agent2.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# Replace inf and -inf with NaN
df_validation.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
df_agent1.dropna(inplace=True)
```

```python
df_agent2.dropna(inplace=True)
```

```python
df_validation.dropna(inplace=True)
```

# LabelEncoder       

```python
from sklearn.preprocessing import LabelEncoder

atributo = 'Label'
le = LabelEncoder()
df_agent1[atributo] = le.fit_transform(df_agent1[atributo].values)
df_agent2[atributo] = le.fit_transform(df_agent2[atributo].values)
df_validation[atributo] = le.fit_transform(df_validation[atributo].values)
```

```python
print(le.classes_)
```

# Divis√£o entre previsores e classe

```python
agent1_y = df_agent1['Label']
agent1_x = df_agent1.drop('Label', axis=1)
agent2_y = df_agent2['Label']
agent2_x = df_agent2.drop('Label', axis=1)
validation_y = df_validation['Label']
validation_x = df_validation.drop('Label', axis=1)
```

# Normaliza√ß√£o

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler        = MinMaxScaler()
x_norm               = minmax_scaler.fit_transform(agent1_x)
agent1_x          = pd.DataFrame(x_norm)
x_norm               = minmax_scaler.fit_transform(agent2_x)
agent2_x          = pd.DataFrame(x_norm)
x_norm               = minmax_scaler.fit_transform(validation_x)
validation_x      = pd.DataFrame(x_norm)
```

# Salvar bases de dados

```python
# agent1_x = pd.DataFrame(agent1_pca)
agent1_y = pd.DataFrame(agent1_y)
# agent2_x = pd.DataFrame(agent1_pca)
agent2_y = pd.DataFrame(agent2_y)
# validation_x = pd.DataFrame(validation_pca)
validation_y = pd.DataFrame(validation_y)
```

```python
agent1_x = agent1_x.drop_duplicates()
agent2_x = agent2_x.drop_duplicates()
validation_x = validation_x.drop_duplicates()
```

```python
# Convert only float64 columns to float32
agent1_x = agent1_x.apply(lambda x: x.astype('float32') if x.dtype == 'float64' or x.dtype == 'int64' else x)
agent1_y = agent1_y.apply(lambda x: x.astype('float32') if x.dtype == 'float64' or x.dtype == 'int64' else x)
agent2_x = agent2_x.apply(lambda x: x.astype('float32') if x.dtype == 'float64' or x.dtype == 'int64' else x)
agent2_y = agent2_y.apply(lambda x: x.astype('float32') if x.dtype == 'float64' or x.dtype == 'int64' else x)
validation_x = validation_x.apply(lambda x: x.astype('float32') if x.dtype == 'float64' or x.dtype == 'int64' else x)
validation_y = validation_y.apply(lambda x: x.astype('float32') if x.dtype == 'float64' or x.dtype == 'int64' else x)
```
