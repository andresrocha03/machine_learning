```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import sklearn
```

# Importar dados

```python
data = pd.read_csv('data/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv')
```

```python
data.sample(300)
```

# Tratamento de dados

```python
data.info()
```

```python
data.isnull().sum()
```

```python
# Replace inf and -inf with NaN
data.replace([np.inf, -np.inf], np.nan, inplace=True)
print(data)
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
data.head()
```

# Divisão entre previsores e classe

```python
df_train_y = data[' Label']
df_train_x = data.drop(' Label', axis=1)
```

```python
for i,r in data.iterrows():
    if data.at[i, ' Label'] == "BENIGN":
        data.at[i, ' Label'] = 1
    else:
        data.at[i, ' Label'] = 0
        
```

# Normalização

```python
from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()
x_norm        = minmax_scaler.fit_transform(df_train_x)
data          = pd.DataFrame(x_norm)
data.head()
```
